We only need to find ONE context that they appear in, and we gradually add context, we start from one word

```python
sampled_passage_indicies = set()

# we can't control the exact window size because we are not running BPE here...
# I think it's ok...
# a rough distance is good enough.

word_to_sampled_passage = {}

for word in sampled_words:
    ind = random.choice(list(vocab_loc[word]))
    while ind in sampled_passage_indicies:
        ind = random.choice(list(vocab_loc[word]))
        # possibility of an infinite loop here...but it's ok
    sampled_passage_indicies.add(ind)
    word_to_sampled_passage[word] = ind
```

```python
# we generate context size 5, 11, 17
def generate_context(word, passage):
    gradual_passages = []
    alltext = [t.text for t in passage]
    w_idx = alltext.index(word)
    for window_size in range(1, len(passage), 2):
        # we start at the word intself, and then we gradually add 
        window_size = window_size - 1
        if window_size == 0:
            gradual_passages.append([passage[w_idx]])
        else:
            remain_wind_size = window_size
            start_idx = w_idx
            end_idx = w_idx
            while remain_wind_size != 0:
                if start_idx > 0:
                    start_idx = start_idx - 1
                    remain_wind_size -= 1
                if end_idx < len(passage):
                    end_idx = end_idx + 1
                    remain_wind_size -= 1
            
            gradual_passages.append(passage[start_idx:end_idx+1])
            
    return gradual_passages
```

```python
from sklearn.manifold import TSNE

words = ['praised', 'surfaced', 'dish']

def tsne_plot(words):
    "Creates and TSNE model and plots it"
    labels = []
    tokens = []

    for word in words:
        tokens.extend(word_to_embeddings[word])
        labels.extend([word + '_{}'.format(i) for i in range(len(word_to_embeddings[word]))])
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(tokens)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        if 'praised' in labels[i]:
            plt.scatter(x[i],y[i], c='red')
        elif 'surfaced' in labels[i]:
            plt.scatter(x[i],y[i], c='blue')
        else:
            plt.scatter(x[i],y[i], c='green')
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```
